```python
# World Happiness Report Data 2018
# Dataset link: https://www.kaggle.com/njlow1202/world-happiness-report-data-2018

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_path = './WorldHappiness2018_Data.csv'
data = pd.read_csv(data_path)
```

```python
# Top-10 happiest country

top_10_rank = data['Rank'].head(10).tolist()
top_10_country = data['Country'].head(10).tolist()
top_10_score = data['Score'].head(10).tolist()

top_10_happiest = {'Rank': top_10_rank,
                   'Country': top_10_country,
                   'Score': top_10_score}
pd.DataFrame(top_10_happiest)
```

```python
# GDP per capita: Economic contribution to happiness score
# Social support: Social contribution to happiness score
# Healthy life expectancy: Health contribution to happiness score
# Freedom to make life choices: Freedom contribution to happiness score
# Generosity: Generosity contribution to happiness score
# Perceptions of corruption: Trustworthiness contribution to happiness score
# Residual: Portion of happiness score that is not attributed to any of the listed categories

x = ['GDP per capita',
     'Social support',
     'Healthy life expectancy',
     'Freedom to make life choices',
     'Generosity',
     'Perceptions of corruption',
     'Residual']

# The 1st happiest country: Finland

y = [data['GDP_Per_Capita'][0],
     data['Social_Support'][0],
     data['Healthy_Life_Expectancy'][0],
     data['Freedom_To_Make_Life_Choices'][0],
     data['Generosity'][0],
     data['Perceptions_Of_Corruption'][0],
     data['Residual'][0]]

plt.subplots(figsize=(17.5, 7.5))
plt.title('The 1st happiest country: Finland')
plt.bar(x, y, width=0.5)
plt.grid()
plt.show()

# The 2nd happiest country: Norway

y = [data['GDP_Per_Capita'][1],
     data['Social_Support'][1],
     data['Healthy_Life_Expectancy'][1],
     data['Freedom_To_Make_Life_Choices'][1],
     data['Generosity'][1],
     data['Perceptions_Of_Corruption'][1],
     data['Residual'][1]]

plt.subplots(figsize=(17.5, 7.5))
plt.title('The 2nd happiest country: Norway')
plt.bar(x, y, width=0.5)
plt.grid()
plt.show()

# The 3rd happiest country: Denmark

y = [data['GDP_Per_Capita'][2],
     data['Social_Support'][2],
     data['Healthy_Life_Expectancy'][2],
     data['Freedom_To_Make_Life_Choices'][2],
     data['Generosity'][2],
     data['Perceptions_Of_Corruption'][2],
     data['Residual'][2]]

plt.subplots(figsize=(17.5, 7.5))
plt.title('The 3rd happiest country: Denmark')
plt.bar(x, y, width=0.5)
plt.grid()
plt.show()
```
